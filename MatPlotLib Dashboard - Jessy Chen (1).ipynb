## Line Chart

## 3 Minute Story

#### Pie Chart
With my pie chart, I really wanted to analyze the ethnicity of the mother with relation to their choice of baby names. I noticed that some of the mother's ethnicities were spelled differently when it was clearly speaking to the same ethnicity, so first I cleaned up the data to get rid of spelling errors. Then I created a new dataframe that just pulled in the two columns I was most interested in - Count and the Ethnicity of the Mother. In order to aggregate all the babies born to the different ethnicities of the mothers, I used the groupby function. So now instead of 27,491 rows to look through, I had only 4. From there it was a lot easier to visualize all that information in a pie chart. It showed that the White Non-Hispanics and Hispanics composed of more than half of the ethnicities of mothers who named their baby with one of the popular baby names. This information could be really useful because we can use this data if we have baby products and we know that we have to be more sensitive in terms of marketing to certain ethinicities.

#### Bar Chart
For my bar chart, I wanted to look at the top ranking girl names in 2016. However, I noticed that some of the names in the data were either all uppercase or the first letter in the name was uppercase but everything after was lowercase. So I cleaned up the data by just making all the names in the 'Childs First Names' column uppercase. Then I filtered so that the gender was female and the year was 2016. Then because I wanted to look at only the top ranking girl names, I filtered to show me the names with ranks of less than 3. Then I performed a groupby function so that I could get the sum of the count of babies born by name. Now that the data was a lot nicer to look at, I created a bar chart which showed the names of the baby girls on the x-axis and the count of girls with that name on the y-axis. This showed that Olivia was the most popular girl name. 

#### Scatter Plot
For my scatter plot, I wanted to see the total amount of babies born by the rank of their names. That way I could see an aggregate of all the names over the years and the ranks. Since in some years, the total number of babies born might climb dramatically and the ranks of names might not change, it can affect how the scatterplot looks like. So first I created a new dataframe that filtered for all boys and did a groupby function to get the sum of the babies born by rank. I did the same for baby girls. Then I used these 2 dataframes to create a scatter plot of the number both girls and boys born by the rank of their names. It was really interesting to see how the plots look like they're going in a downward right angle, but once you get to the the 40th rank for girls and 50th rank for boys, there are fluctuations in the amount of babies born with that name. It could have to do with a baby boom during certain years since this is an aggregate of all the years from the Popular Baby Names data source.

#### Line Chart
For my line chart, I was interested in seeing the total amount of babies born with popular names by year and then the breakout between boys and girls. So I did a groupby function by gender and year of birth to get the total sum of the count of babies. Since the resulting dataframe didn't really produce an easy to read line chart, I used the unstack function to change my dataframe so that the year of birth was the index and the gender was broken out in the columns. Once I had my new dataframe, the new line chart was much easier to read especially since the colors of the lines name it much easier to distinguish that boys were named more often with popular names across all the years. This data can be really useful especially for teachers on the first day of school since they can expect girls to have less popular or more unique names than boys, so they can expect to struggle more pronouncing the girls names than boys names.

#### Histogram
Finally, for my histogram, I wanted to see the distribution of the ranks by the number of names with that rank. Since there were some columns that I didn't need, I decided to drop the nucessary columns when I created a new dataframe. Once I created the histogram, I was very concerned with how the histogram looked since I didn't want the number of bins to be overwhelming. So I decided that 5 bins was a nice round number. Then to distinguish between the bins, I made the edgecolors of the bars in the histogram black. It's through this histogram that shows that the average rank of these popular baby names is somewhere between the 60th and 80th percentile.
